```python
new_pressure = []
new_temperature = []
for i in range(len(data['Pressure']) - 1):
    new_t =(data.iloc[i]['Temperature'] + data.iloc[i+1]['Temperature']) / 2
    new = (data.iloc[i]['Pressure'] + data.iloc[i+1]['Pressure']) / 2
    new_pressure.append(new)
    new_temperature.append(new_t)
```
